# 전처리(3) - 재료, 양념 합치기

```python
df_total = df_go.copy()

# 1️⃣ 재료 + 양념 합치기
def join_ingr_season(row):
    a = str(row.get("재료", "") if pd.notna(row.get("재료", np.nan)) else "").strip()
    b = str(row.get("양념", "") if pd.notna(row.get("양념", np.nan)) else "").strip()
    s = (a + " " + b).strip()
    return " ".join(s.split())  # 다중 공백 정리

df_total["재료양념_raw"] = df_total.apply(join_ingr_season, axis=1)

# 2️⃣ 레시피 단위로 통합 (중복 제거 + 순서 보존)
def uniq_tokens_preserve_order(series):
    seen = set()
    out = []
    for text in series.astype(str):
        if not text or text == "nan":
            continue
        for tok in text.split():
            if tok and tok not in seen:
                seen.add(tok)
                out.append(tok)
    return " ".join(out).strip()

df_recipe = (
    df_total
    .dropna(subset=["레시피일련번호", "요리명"])
    .groupby(["레시피일련번호", "요리명"], as_index=False)
    .agg(재료통합=("재료양념_raw", uniq_tokens_preserve_order))
)

# 3️⃣ 다중 공백 제거
df_recipe["재료통합"] = df_recipe["재료통합"].str.replace(r"\s+", " ", regex=True).str.strip()

# 4️⃣ 토큰 리스트 컬럼 추가 (Word2Vec용)
df_recipe["재료토큰"] = df_recipe["재료통합"].apply(lambda x: x.split())

print("✅ 레시피 단위 재료+양념 통합 + 토큰화 완료")
print(f"총 {len(df_recipe):,}개의 고유 레시피로 축약되었습니다.")
```

# 재료 대체

```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
from collections import defaultdict, Counter
import itertools

# ============================================
# 공통: DataFrame 컬럼 이름 상수
# ============================================
COL_RECIPE_ID = "레시피일련번호"
COL_DISH_NAME = "요리명"
COL_ING_LIST  = "재료토큰"
COL_METHOD    = "요리방법별명"
COL_CAT       = "요리종류별명_세분화"
```

```python
def build_index_maps(df_recipe):
    """
    반환:
      - name_to_ids          : 요리명 -> 레시피ID 리스트
      - method_map           : 레시피ID -> 조리방법(볶음/찌개 등)
      - recipes_by_ingredient: 재료명 -> 이 재료가 들어간 레시피ID 리스트
      - cat_map              : 레시피ID -> 요리종류별명_세분화(밥/면/소스 등)
    """
    name_to_ids = defaultdict(list)
    method_map = {}
    recipes_by_ingredient = defaultdict(list)
    cat_map = {}   # ✅ 추가

    for _, row in df_recipe.iterrows():
        rid    = row[COL_RECIPE_ID] # 레시피일련번호
        name   = row[COL_DISH_NAME] # 요리명
        method = row[COL_METHOD]    # 요리방법별명
        ings   = row[COL_ING_LIST]  # 재료토큰 
        cat    = row[COL_CAT]       # 요리종류별명_세분화

        name_to_ids[name].append(rid)
        method_map[rid] = method
        cat_map[rid]    = cat       # ✅ 레시피ID -> 세분화 카테고리

        for ing in ings:
            recipes_by_ingredient[ing].append(rid)

    return name_to_ids, method_map, recipes_by_ingredient, cat_map
```

## 요리종류별명_세분화

```python
# ============================================
# 재료별 요리종류별명_세분화 카운트 구축
# ============================================
def build_ing_category_counts(df_recipe):
    """
    각 재료가 어떤 '요리종류별명_세분화'(밥/면/소스/양념/...)에서
    얼마나 자주 등장했는지 Counter로 저장.

    반환: ing_cat_counts: dict
        재료명 -> Counter({카테고리명: 등장횟수, ...})
    """
    ing_cat_counts = defaultdict(Counter)

    for _, row in df_recipe.iterrows():
        cat = row[COL_CAT]          # 예: '밥', '면', '소스', ...
        ings = row[COL_ING_LIST]    # 재료 리스트
        for ing in ings:
            ing_cat_counts[ing][cat] += 1

    return ing_cat_counts
```

```python
# ============================================
# 재료-카테고리 호환도 스코어
# ============================================
def ing_category_score(ingredient, current_cat, ing_cat_counts, min_count=5):
    """
    ingredient : 후보 재료명
    current_cat: 현재 레시피의 요리종류별명_세분화 (예: '밥')
    ing_cat_counts: build_ing_category_counts 결과

    반환: 0~1 사이 스코어
      - 1에 가까울수록 "이 재료는 현재 카테고리에서 자주 쓰인다"
      - 0에 가까울수록 "이 카테고리에서는 거의 안 쓰인다"
      - 등장 횟수 자체가 적으면 0.5(중립)로 처리
    """
    cnts = ing_cat_counts.get(ingredient)
    if not cnts:
        return 0.5  # 정보 없음 → 중립

    total = sum(cnts.values())
    if total < min_count:
        return 0.5  # 너무 희귀한 재료 → 중립

    return cnts[current_cat] / total  # 이 카테고리에서의 비율 (0~1)
```

```python
# ============================================================
# 1. 요리명으로 레시피 목록 찾기 (input() 기반)
# ============================================================

def interactive_find_recipes_by_name(df_recipe, name_to_ids, max_results=10):
    """
    사용자에게 '요리명'을 입력받아,
      - 해당 요리명의 레시피 후보 목록을 보여주고
      - 사용자에게 번호를 선택하게 한 뒤
      - 선택된 레시피의 '레시피일련번호'를 반환하는 함수.
    """
    dish_name = input("🍽 대체를 원하는 요리명을 입력하세요 (예: 파스타): ").strip()
    if not dish_name:
        print("❌ 요리명이 비어 있습니다.")
        return None

    # 1) 요리명 정확 일치 우선
    if dish_name in name_to_ids:
        candidate_ids = name_to_ids[dish_name]
        sub = df_recipe[df_recipe[COL_RECIPE_ID].isin(candidate_ids)].copy()
    else:
        # 2) 없으면 부분 일치 검색 (contains)
        sub = df_recipe[df_recipe[COL_DISH_NAME].str.contains(dish_name, na=False)].copy()
        if sub.empty:
            print(f"❌ '{dish_name}'에 해당하는 레시피를 찾을 수 없습니다.")
            return None

    sub = sub.head(max_results)

    print(f"\n🧾 검색 결과 (최대 상위 {max_results}개)")
    print("-" * 70)

    for i, (_, row) in enumerate(sub.iterrows(), start=1):
        rid    = row[COL_RECIPE_ID]
        name   = row[COL_DISH_NAME]
        method = row[COL_METHOD]
        ings   = row[COL_ING_LIST]
        cat    = row[COL_CAT]

        print(f"[{i}] 레시피ID={rid} | 요리명={name} | 조리방법={method} | 요리종류별명_세분화={cat}")
        print("    재료:", ", ".join(ings))
        print("-" * 70)

    idx = input("➡ 사용할 레시피 번호를 선택하세요 (예: 1): ").strip()
    if not idx.isdigit():
        print("❌ 숫자를 입력해야 합니다.")
        return None
    idx = int(idx)

    if not (1 <= idx <= len(sub)):
        print("❌ 범위를 벗어난 번호입니다.")
        return None

    chosen_row = sub.iloc[idx - 1]
    chosen_id  = chosen_row[COL_RECIPE_ID]
    chosen_cat = chosen_row[COL_CAT]

    print(
    f"\n✅ 선택된 레시피ID: {chosen_id} "
    f"(요리명: {chosen_row[COL_DISH_NAME]}, 방법: {chosen_row[COL_METHOD]}, 종류: {chosen_cat})"
    )
    print("   재료:", ", ".join(chosen_row[COL_ING_LIST]))
    print("=" * 70)
    return chosen_id
```

```python
# ============================================================
# 2-1. Doc2Vec + Word2Vec + 카테고리 기반 단일 재료 대체 함수
# ============================================================

def substitute_with_method(
    w2v_model,
    d2v_model,
    df_recipe,
    recipes_by_ingredient,
    method_map,
    ing_cat_counts,   # ✅ 추가: 재료별 카테고리 카운트
    cat_map,          # ✅ 추가: 레시피ID -> 요리종류별명_세분화
    recipe_id,
    target_ing,
    topn_base=50,
    topn_final=10,
    alpha=0.7,   # 가중치 고정
):
    """
    [단일 재료 대체]

    - Word2Vec: target 재료와 유사한 재료 후보를 먼저 뽑고
    - Doc2Vec: 현재 레시피(재료 + 조리방법)와의 문맥/조리방법 적합도
    - 요리종류별명_세분화(밥/면/소스/양념...): 해당 카테고리에서 자주 쓰이는 재료인지 보정
    - 최종 점수 = (0.7 * 재료유사도 + 0.3 * 문맥/조리방법) * 카테고리보정

      카테고리보정은 (0.5 + 0.5 * cat_score) 형태로,
      - cat_score=1.0 → 1.0배 (완전 호환)
      - cat_score=0.0 → 0.5배 (절반으로 깎기)
    """
    
    # 현재 레시피 정보
    row = df_recipe[df_recipe[COL_RECIPE_ID] == recipe_id]
    if row.empty:
        print(f"❌ 레시피일련번호 {recipe_id} 를 df_recipe에서 찾을 수 없습니다.")
        return pd.DataFrame()

    row = row.iloc[0]
    current_method = row[COL_METHOD]
    context_ings   = row[COL_ING_LIST]

    # ✅ 현재 레시피의 요리종류 카테고리 (밥/면/소스/양념 등)
    current_cat = cat_map.get(recipe_id, None)
    # cat_map을 안 쓰고 바로 df_recipe에서 읽고 싶다면:
    # current_cat = row[COL_CAT]

    if target_ing not in context_ings:
        print(f"⚠️ '{target_ing}'는 레시피ID {recipe_id}의 재료 목록에 없습니다.")
        return pd.DataFrame()

    tag = f"recipe_{recipe_id}"
    if tag not in d2v_model.dv:
        print(f"⚠️ Doc2Vec 모델에서 {tag} 벡터를 찾을 수 없습니다.")
        return pd.DataFrame()

    vec_recipe = d2v_model.dv[tag]

    # 1) Word2Vec 기반 기본 후보 (재료 자체 유사도)
    if target_ing not in w2v_model.wv:
        print(f"⚠️ Word2Vec 모델에 '{target_ing}' 임베딩이 없습니다.")
        return pd.DataFrame()

    base_candidates = [
        (ing, sim)
        for ing, sim in w2v_model.wv.most_similar(target_ing, topn=topn_base)
        if ing not in context_ings  # 이미 레시피에 있는 재료는 제외
    ]

    if not base_candidates:
        print("❌ Word2Vec 기반 후보 재료가 없습니다.")
        return pd.DataFrame()

    # 2) Doc2Vec 기반 문맥/조리방법 적합도 계산
    def cos_sim(a, b):
        a = np.asarray(a)
        b = np.asarray(b)
        denom = (np.linalg.norm(a) * np.linalg.norm(b) + 1e-9)
        return float(np.dot(a, b) / denom)

    results = []
    for cand, sim_ing in base_candidates:
        # (a) Doc2Vec 문맥 점수: 후보 재료가 자주 쓰인 "같은 조리방법" 레시피들과의 유사도
        rid_list = recipes_by_ingredient.get(cand, [])
        same_method_ids = [rid for rid in rid_list if method_map.get(rid) == current_method]

        if same_method_ids:
            sims = []
            for rid in same_method_ids:
                cand_tag = f"recipe_{rid}"
                if cand_tag in d2v_model.dv:
                    sims.append(cos_sim(vec_recipe, d2v_model.dv[cand_tag]))
            sim_ctx = float(np.mean(sims)) if sims else 0.0
        else:
            sim_ctx = 0.0

        # (b) 기본 점수 (W2V + Doc2Vec)
        raw_score = alpha * sim_ing + (1 - alpha) * sim_ctx

        # (c) ✅ 카테고리 호환도 스코어 (요리종류별명_세분화)
        if current_cat is not None:
            cat_score = ing_category_score(cand, current_cat, ing_cat_counts)
        else:
            cat_score = 0.5  # 정보 없으면 중립

        # 0.5 ~ 1.0 배 사이로 조정
        final_score = raw_score * (0.5 + 0.5 * cat_score)

        results.append((cand, sim_ing, sim_ctx, final_score))

    if not results:
        print("❌ 조리방법/카테고리를 고려했을 때 유효한 후보가 없습니다.")
        return pd.DataFrame()

    df_res = pd.DataFrame(
        results,
        columns=["후보재료", "재료유사도_w2v", "문맥/조리방법_적합도_d2v", "최종점수"]
    )
    df_res = df_res.sort_values("최종점수", ascending=False).head(topn_final).reset_index(drop=True)
    return df_res
```

```python
# ============================================================
# 2-2. 다중 재료 대체 함수 (Word2Vec + Doc2Vec + 조리방법 + 카테고리 + 조합제한)
# ============================================================

def multi_substitute_with_method(
    w2v_model,
    d2v_model,
    df_recipe,
    recipes_by_ingredient,
    method_map,
    ing_cat_counts,   # ✅ 추가: 재료별 카테고리 카운트
    cat_map,          # ✅ 추가: 레시피ID -> 요리종류별명_세분화
    recipe_id,
    targets,
    topn_each=5,
    result_topn=10,
    alpha=0.7,
    max_combos=5000,
):
    """
    [다중 재료 대체]

    여러 개의 target 재료를 동시에 바꾸고 싶을 때:
      - 각 타겟별로 Word2Vec 상위 유사 재료를 후보로 뽑고
      - 후보 조합(카트esian product)을 만들되, max_combos를 넘지 않게 제한하고
      - 각 조합에 대해
          * Word2Vec: 원래 레시피 vs 대체 후 레시피 벡터 유사도
          * Doc2Vec: 각 후보 재료가 현재 레시피(조리방법 포함)에 어울리는지 평균 점수
          * Category: 해당 조합의 재료들이 현재 '요리종류별명_세분화'(밥/면/소스...)에 얼마나 잘 맞는지
        를 계산한 뒤
      - 최종 점수 = (0.7 * w2v_score + 0.3 * doc2vec_score) * 카테고리보정
    """
    row = df_recipe[df_recipe[COL_RECIPE_ID] == recipe_id]
    if row.empty:
        print(f"❌ 레시피일련번호 {recipe_id} 를 df_recipe에서 찾을 수 없습니다.")
        return []

    row = row.iloc[0]
    context_ings   = row[COL_ING_LIST]
    current_method = row[COL_METHOD]
    # ✅ 현재 레시피의 요리종류 카테고리 (밥/면/소스/...)
    current_cat    = cat_map.get(recipe_id, None)

    # context(타겟 제외)의 W2V 평균 벡터
    context_vecs = [w2v_model.wv[w] for w in context_ings
                    if w in w2v_model.wv and w not in targets]
    if not context_vecs:
        print("⚠️ 유효한 문맥 재료가 없습니다 (타겟 제외).")
        return []

    base_vec = np.mean(context_vecs, axis=0)

    # 원래 타겟들의 평균 벡터 (원래 레시피 표현)
    target_vecs = [w2v_model.wv[w] for w in targets if w in w2v_model.wv]
    if not target_vecs:
        print("⚠️ 타겟 재료 중 Word2Vec 임베딩이 없는 것이 있습니다.")
        return []

    target_vec_mean     = np.mean(target_vecs, axis=0)
    original_recipe_vec = base_vec + target_vec_mean

    # Doc2Vec에서 현재 레시피 벡터
    tag = f"recipe_{recipe_id}"
    if tag not in d2v_model.dv:
        print(f"⚠️ Doc2Vec 모델에서 {tag} 벡터를 찾을 수 없습니다.")
        return []

    vec_recipe_doc = d2v_model.dv[tag]

    # 1) 각 타겟 재료별 후보 pool (Word2Vec 기반)
    candidate_pools = {}
    for t in targets:
        if t not in w2v_model.wv:
            candidate_pools[t] = []
            continue

        similar = w2v_model.wv.most_similar(t, topn=topn_each * 3)
        cand_list = [w for w, _ in similar if w not in context_ings]  # 이미 레시피에 있는 재료 제외
        candidate_pools[t] = cand_list[:topn_each]

    if not all(candidate_pools.values()):
        print("⚠️ 일부 타겟 재료에 대한 후보가 없습니다.")
        return []

    # 2) 조합 수 제한
    sizes = [len(pool) for pool in candidate_pools.values()]
    est_combos = int(np.prod(sizes))
    if est_combos > max_combos:
        m = len(targets)
        k = int(max_combos ** (1.0 / m))
        k = max(1, min(k, topn_each))

        print(f"⚠️ 예상 조합 수가 {est_combos}개로 너무 큽니다.")
        print(f"   각 타겟당 후보 수를 최대 {k}개로 자동 축소합니다. (max_combos={max_combos})")

        for t in targets:
            candidate_pools[t] = candidate_pools[t][:k]

        sizes = [len(pool) for pool in candidate_pools.values()]
        est_combos = int(np.prod(sizes))
        print(f"   축소 후 예상 조합 수: {est_combos}개")

    def cos_doc(a, b):
        a = np.asarray(a)
        b = np.asarray(b)
        denom = (np.linalg.norm(a) * np.linalg.norm(b) + 1e-9)
        return float(np.dot(a, b) / denom)

    # Doc2Vec 기반 재료별 "조리방법/문맥 적합도" 캐시
    method_fit_cache = {}

    def get_method_fit(ingredient):
        if ingredient in method_fit_cache:
            return method_fit_cache[ingredient]

        rid_list = recipes_by_ingredient.get(ingredient, [])
        same_method_ids = [rid for rid in rid_list if method_map.get(rid) == current_method]

        if same_method_ids:
            sims = []
            for rid in same_method_ids:
                t = f"recipe_{rid}"
                if t in d2v_model.dv:
                    sims.append(cos_doc(vec_recipe_doc, d2v_model.dv[t]))
            score = float(np.mean(sims)) if sims else 0.0
        else:
            score = 0.0

        method_fit_cache[ingredient] = score
        return score

    # 3) 모든 조합에 대해 점수 계산
    results = []
    pools_in_order = [candidate_pools[t] for t in targets]

    for combo in itertools.product(*pools_in_order):
        # (a) Word2Vec 기반 레시피 유사도
        combo_vecs = [w2v_model.wv[w] for w in combo if w in w2v_model.wv]
        if not combo_vecs:
            continue
        combo_vec_mean = np.mean(combo_vecs, axis=0)
        new_recipe_vec = base_vec + combo_vec_mean

        w2v_score = float(
            cosine_similarity([original_recipe_vec], [new_recipe_vec])[0][0]
        )

        # (b) Doc2Vec 기반 조리방법/문맥 적합도
        method_scores = [get_method_fit(w) for w in combo]
        method_score  = float(np.mean(method_scores)) if method_scores else 0.0

        # (c) ✅ 카테고리 호환도 (조합 내 재료 평균)
        if current_cat is not None:
            cat_scores = [ing_category_score(w, current_cat, ing_cat_counts) for w in combo]
            cat_score  = float(np.mean(cat_scores)) if cat_scores else 0.5
        else:
            cat_score = 0.5  # 카테고리 정보 없으면 중립

        # (d) 최종 점수 계산
        raw_score   = alpha * w2v_score + (1 - alpha) * method_score
        # 0.5~1.0 배 사이 보정: 카테고리에 잘 맞을수록 점수 유지/상승, 아니면 깎임
        final_score = raw_score * (0.5 + 0.5 * cat_score)

        results.append((combo, w2v_score, method_score, final_score))

    if not results:
        print("❌ 유효한 대체 조합을 만들 수 없습니다.")
        return []

    results = sorted(results, key=lambda x: x[3], reverse=True)[:result_topn]
    return results
```

```python
# ============================================================
# 3. 통합 인터페이스 (input 기반) - 레시피ID + 타겟 재료
# ============================================================

def run_recipe_substitution(w2v_model, d2v_model, df_recipe):
    """
    전체 파이프라인을 input() 기반으로 실행:

      1) 요리명 입력 -> 레시피 후보 리스트
      2) 레시피 번호 선택 -> 레시피ID 확정
      3) 모드 선택: 단일(1) / 다중(2)
      4) 선택된 모드에 따라 대체 재료(조합) 추천
    """
    print("==== 🔄 레시피 기반 재료 대체 시스템 ====")

    # 인덱스/매핑 + 카테고리 매핑 생성
    name_to_ids, method_map, recipes_by_ingredient, cat_map = build_index_maps(df_recipe)

    # 재료별 카테고리 등장 카운트 (한 번만 만들고 재사용)
    ing_cat_counts = build_ing_category_counts(df_recipe)


    # 1) 요리명으로 레시피 선택
    recipe_id = interactive_find_recipes_by_name(df_recipe, name_to_ids)
    if recipe_id is None:
        return

    row = df_recipe[df_recipe[COL_RECIPE_ID] == recipe_id].iloc[0]
    context_ings = row[COL_ING_LIST]
    method       = row[COL_METHOD]

    print("\n📌 선택된 레시피 정보")
    print(f"   레시피ID: {recipe_id}")
    print(f"   요리명  : {row[COL_DISH_NAME]}")
    print(f"   조리방법: {method}")
    print(f"   재료목록: {', '.join(context_ings)}")
    print("=" * 70)

    mode = input("단일 재료 대체(1) / 다중 재료 대체(2): ").strip()

    # ---- 단일 모드 ----
    if mode == "1":
        target_ing = input("🎯 바꾸고 싶은 재료명을 입력하세요 (위 재료 목록 중 하나): ").strip()
        if target_ing not in context_ings:
            print(f"⚠️ '{target_ing}'는 이 레시피의 재료 목록에 없습니다.")
            return

        df_res = substitute_with_method(
            w2v_model=w2v_model,
            d2v_model=d2v_model,
            df_recipe=df_recipe,
            recipes_by_ingredient=recipes_by_ingredient,
            method_map=method_map,
            ing_cat_counts=ing_cat_counts,
            cat_map=cat_map, 
            recipe_id=recipe_id,
            target_ing=target_ing,
            topn_base=50,
            topn_final=10,
            alpha=0.7
        )

        if df_res.empty:
            print("❌ 추천할 대체 재료가 없습니다.")
            return

        print(f"\n✅ '{target_ing}' 대체 후보 (재료 유사도 + 조리방법 고려):")
        print("-" * 70)
        for i, r in df_res.iterrows():
            print(
                f"[{i+1}] 후보재료: {r['후보재료']:<15s} | "
                f"재료유사도={r['재료유사도_w2v']:.3f} | "
                f"문맥/조리방법={r['문맥/조리방법_적합도_d2v']:.3f} | "
                f"최종점수={r['최종점수']:.3f}"
            )
        print("-" * 70)
        print("🎉 추천 완료!")
        return

    # ---- 다중 모드 ----
    elif mode == "2":
        tgt_input = input("🎯 바꾸고 싶은 재료들을 공백/쉼표로 입력하세요 (예: 양파 대파): ").strip()
        targets = [t.strip() for t in tgt_input.replace(',', ' ').split() if t.strip()]

        if not targets:
            print("❌ 타겟 재료가 입력되지 않았습니다.")
            return

        invalid = [t for t in targets if t not in context_ings]
        if invalid:
            print(f"⚠️ 다음 재료는 이 레시피에 없습니다: {', '.join(invalid)}")
            return

        combos = multi_substitute_with_method(
            w2v_model=w2v_model,
            d2v_model=d2v_model,
            df_recipe=df_recipe,
            recipes_by_ingredient=recipes_by_ingredient,
            method_map=method_map,
            ing_cat_counts=ing_cat_counts,
            cat_map=cat_map, 
            recipe_id=recipe_id,
            targets=targets,
            topn_each=5,
            result_topn=10,
            alpha=0.7,
            max_combos=5000,
        )

        if not combos:
            print("❌ 추천할 대체 조합이 없습니다.")
            return

        print("\n✅ 다중 대체 후보 (재료 + 조리방법 고려):")
        print("-" * 90)
        for i, (combo, w2v_score, m_score, final_score) in enumerate(combos, start=1):
            combo_text = ", ".join(combo)
            print(
                f"[{i}] {combo_text:<40s} | "
                f"w2v={w2v_score:.3f} | "
                f"doc2vec={m_score:.3f} | "
                f"최종={final_score:.3f}"
            )
        print("-" * 90)
        print("🎉 다중 대체 추천 완료!")
        return

    else:
        print("❌ 잘못된 입력입니다. 1 또는 2를 선택하세요.")
        return
```

# 단일/다중 재료 대체

```python
import pandas as pd
import numpy as np
import itertools
from sklearn.metrics.pairwise import cosine_similarity

# -----------------------------
# 단일 재료 대체 함수
# -----------------------------
def substitute_in_context(model, target, context, topn=10):
    """
    target: 바꾸려는 재료(str)
    context: 현재 함께 쓰이는 재료 list (문맥)
    """
    context_vecs = [model.wv[w] for w in context if w in model.wv and w != target]
    if not context_vecs:
        print("⚠️ 유효한 문맥 재료가 없습니다.")
        return pd.DataFrame()
    context_mean = np.mean(context_vecs, axis=0)

    words = list(model.wv.index_to_key)
    all_vecs = np.array([model.wv[w] for w in words])
    target_vec = model.wv[target]

    recipe_vec = context_mean + target_vec
    scores = cosine_similarity([recipe_vec], all_vecs)[0]

    df_sim = pd.DataFrame({"재료": words, "유사도": scores})
    
    # ✅ 이미 레시피에 있는 재료들 제외
    df_sim = df_sim[~df_sim["재료"].isin(context)]
    df_sim = df_sim.sort_values("유사도", ascending=False)
    
    return df_sim.head(topn)

# -----------------------------
# 다중 재료 대체 함수
# -----------------------------
def fast_multi_substitute(model, targets, context, topn_each=20, result_topn=10):
    """
    model         : 학습된 Word2Vec 모델
    targets       : 바꾸려는 재료 리스트
    context       : 전체 레시피의 재료 리스트
    """
    context_vecs = [model.wv[w] for w in context if w in model.wv and w not in targets]
    if not context_vecs:
        print("⚠️ 유효한 문맥 재료가 없습니다.")
        return []
    base_vec = np.mean(context_vecs, axis=0)

    target_vec = np.mean([model.wv[w] for w in targets if w in model.wv], axis=0)
    original_recipe_vec = base_vec + target_vec

    # 후보 pool 생성
    candidate_pools = {}
    for t in targets:
        if t in model.wv:
            # ✅ 유사한 재료를 가져올 때 이미 레시피에 있는 재료는 제외
            similar = model.wv.most_similar(t, topn=topn_each * 2)  # 여유있게 더 많이 가져옴
            candidate_pools[t] = [
                w for w, _ in similar 
                if w not in context  # 레시피에 없는 재료만
            ][:topn_each]  # 필터링 후 topn_each개만 선택
        else:
            candidate_pools[t] = []

    if not all(candidate_pools.values()):
        print("⚠️ 일부 재료에 대한 후보가 없습니다.")
        return []

    combos = list(itertools.product(*candidate_pools.values()))

    results = []
    for combo in combos:
        combo_vec = np.mean([model.wv[w] for w in combo if w in model.wv], axis=0)
        new_recipe_vec = base_vec + combo_vec
        score = cosine_similarity([original_recipe_vec], [new_recipe_vec])[0][0]
        results.append((combo, score))

    results = sorted(results, key=lambda x: x[1], reverse=True)
    return results[:result_topn]

# -----------------------------
# 통합 인터페이스
# -----------------------------
def run_substitution(model):
    print("==== 🔄 재료 대체 시스템 ====")
    mode = input("단일 대체(1) / 다중 대체(2)를 선택하세요: ").strip()

    # context 입력
    ctx_input = input("현재 요리의 전체 재료를 공백 또는 쉼표로 입력하세요 (예: 간장 마늘 참기름 설탕 대파): ").strip()
    # 공백과 쉼표 모두 처리
    context = [ing.strip() for ing in ctx_input.replace(',', ' ').split() if ing.strip()]

    # ✅ 입력된 전체 재료 확인
    print("\n🧾 입력된 전체 재료 목록:")
    print(", ".join(context))
    print("=" * 60)

    # 실행 분기
    if mode == "1":
        target = input("바꾸고 싶은 재료를 입력하세요 (예: 간장): ").strip()
        
        # 타겟이 context에 없으면 경고
        if target not in context:
            print(f"⚠️ '{target}'는 입력된 재료 목록에 없습니다.")
            return
        
        print(f"\n🎯 선택된 대상 재료: {target}")
        print("-" * 60)

        result = substitute_in_context(model, target, context, topn=10)
        if result is not None and not result.empty:
            print(f"\n✅ '{target}' 대체 후보 (문맥 기반 추천):")
            for _, row in result.iterrows():
                print(f"{row['재료']:<15s}  유사도={row['유사도']:.3f}")
        else:
            print("❌ 추천할 대체 재료가 없습니다.")

    elif mode == "2":
        tgt_input = input("바꾸고 싶은 재료들을 공백 또는 쉼표로 입력하세요 (예: 간장 마늘): ").strip()
        targets = [ing.strip() for ing in tgt_input.replace(',', ' ').split() if ing.strip()]
        
        # 타겟들이 context에 있는지 확인
        invalid_targets = [t for t in targets if t not in context]
        if invalid_targets:
            print(f"⚠️ 다음 재료는 입력된 재료 목록에 없습니다: {', '.join(invalid_targets)}")
            return
        
        print(f"\n🎯 선택된 대상 재료들: {', '.join(targets)}")
        print("-" * 60)

        candidates = fast_multi_substitute(model, targets, context, topn_each=20, result_topn=10)
        
        if candidates:
            print("\n✅ 다중 재료 대체 결과 (문맥 기반 추천):")
            for combo, score in candidates:
                combo_text = ", ".join(combo)
                print(f"{combo_text:<40s} → 유사도={score:.3f}")
        else:
            print("❌ 추천할 대체 조합이 없습니다.")

    else:
        print("❌ 잘못된 입력입니다. 1 또는 2를 선택하세요.")
```
