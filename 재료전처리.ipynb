# 전처리(3) - 재료, 양념 합치기

```python
len(df_go['레시피일련번호'].unique())
```

```python
# 1️⃣ 재료 + 양념 합치기 (기존 동일)
df_total = df_go.copy()

def join_ingr_season(row):
    a = str(row.get("재료", "") if pd.notna(row.get("재료", np.nan)) else "").strip()
    b = str(row.get("양념", "") if pd.notna(row.get("양념", np.nan)) else "").strip()
    s = (a + " " + b).strip()
    return " ".join(s.split())

df_total["재료양념_raw"] = df_total.apply(join_ingr_season, axis=1)

# 2️⃣ 레시피 단위로 통합 (수정됨: 요리종류별명_세분화 포함)
def uniq_tokens_preserve_order(series):
    seen = set()
    out = []
    for text in series.astype(str):
        if not text or text == "nan":
            continue
        for tok in text.split():
            if tok and tok not in seen:
                seen.add(tok)
                out.append(tok)
    return " ".join(out).strip()

# [수정된 부분] .agg() 안에 '요리종류별명_세분화'를 추가했습니다.
df_recipe = (
    df_total
    .dropna(subset=["레시피일련번호", "요리명"])
    .groupby(["레시피일련번호", "요리명"], as_index=False)
    .agg(
        재료통합=("재료양념_raw", uniq_tokens_preserve_order),
        요리종류별명_세분화=("요리종류별명_세분화", "first")  # ✅ 같은 레시피면 카테고리가 같으므로 첫번째 값을 가져옵니다.
    )
)

# 3️⃣ 다중 공백 제거 (기존 동일)
df_recipe["재료통합"] = df_recipe["재료통합"].str.replace(r"\s+", " ", regex=True).str.strip()

# 4️⃣ 토큰 리스트 컬럼 추가 (Word2Vec용) (기존 동일)
df_recipe["재료토큰"] = df_recipe["재료통합"].apply(lambda x: x.split())

print("✅ 레시피 단위 재료+양념 통합 + 토큰화 완료 (카테고리 포함)")
print(f"총 {len(df_recipe):,}개의 고유 레시피로 축약되었습니다.")
display(df_recipe.head()) # 결과 확인
```

# 전처리(4) - 요리방법별명

```python
# 1. df_5의 중복 확인 (눈으로 확인용)
# 같은 레시피 ID인데 요리방법이 여러 개로 적힌 경우가 있는지 봅니다.
dup_check = df_5[df_5.duplicated('레시피일련번호', keep=False)].sort_values('레시피일련번호')
print(f"⚠️ df_5 내 중복된 ID 개수: {len(dup_check)}개")
display(dup_check.head()) 

# 2. df_5 중복 제거 (ID 하나당 첫 번째 요리방법만 남김)
df_5_unique = df_5.drop_duplicates('레시피일련번호', keep='first')
print(f"🧹 중복 제거 후 df_5 크기: {len(df_5_unique):,}개")
```

```python
# 3. 다시 병합 (이제 행 개수가 늘어나지 않을 것입니다)
df_merge = df_recipe.merge(
    df_5_unique[['레시피일련번호', '요리방법별명']],
    on='레시피일련번호',
    how='left'
)

print(f"✅ 최종 병합된 데이터 크기: {len(df_merge):,}개")
```
